```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ecut = np.arange(40, 110, 10)
eng = [-696.891337426, -696.8987943561, -696.9006474331, -696.9010557157, -696.9016054947, -696.9018791467, -696.9019707391]
vac = np.arange(5, 35, 5)
eng_vac = [-1868.81810491, -1868.81813529, -1868.81813157, -1868.81813175, -1868.8181282, -1868.81812958]
eng[:] = [(engs-eng[0])*13.6 for engs in eng]
eng_vac[:] = [(eng_vacs-eng_vac[0])*13600 for eng_vacs in eng_vac]
fig, ax = plt.subplots(1,2, figsize = (8,4))
ax[0].plot(ecut, eng, 'ro-')
ax[0].set_xlabel("energy cutoff (Ry)")
ax[0].set_ylabel("energy (eV)")
ax[0].set_title("ecut test")
ax[1].plot(vac, eng_vac, 'bo-')
ax[1].set_xlabel("vacuum length (A)")
ax[1].set_ylabel("energy (meV)")
ax[1].set_title("vacuum test")
plt.tight_layout()
```
